```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
from keras.utils import pad_sequences
from tensorflow.keras.layers import Bidirectional
```

```python
(x_train,y_train),(x_test,y_test) = imdb.load_data(num_words=10000)
```

```python
x_train = pad_sequences(x_train, padding = 'post', maxlen = 50)
x_test = pad_sequences(x_test, padding = 'post', maxlen = 50)
```

```python
# model architecture
model2 = Sequential()
model2.add(Embedding(input_dim = 10000, output_dim = 2, input_length = 50))
model2.add(Bidirectional(SimpleRNN(32, input_shape = (50,1), return_sequences=False)))
model2.add(Dense(1,activation = 'sigmoid'))

model2.summary()
```

```python
model2.compile(metrics = ['acc'], optimizer = 'adam', loss = 'binary_crossentropy')
```

```python
# RNN model train
# model Training
model2.fit(x_train, y_train, epochs = 5, validation_data = (x_test, y_test))
```

```python
test_data = x_test[0][0:50].reshape(1,-1)
if model2.predict(test_data) < 0.5:
  print('negative review')
else:
  print('positive review')
```
