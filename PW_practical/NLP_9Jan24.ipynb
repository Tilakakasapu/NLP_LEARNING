# embedding and RNN on IMDB dataset

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
from keras.utils import pad_sequences
from tensorflow.keras.layers import Bidirectional
```

```python
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words = 10000)
```

```python
# Data processing
# Step - 1  : Visualization of the data
# Step - 2: Zero Padding
X_train = pad_sequences(X_train, padding = 'post', maxlen = 50)
X_test = pad_sequences(X_test, padding = 'post', maxlen = 50)
```

```python
X_train
```

```python
X_train.shape
```

```python
print(len(X_train[0]))
print(len(X_train[4]))
```

```python
# model architecture
model = Sequential()
model.add(Embedding(input_dim = 10000, output_dim = 2, input_length = 50))
model.add(SimpleRNN(32, input_shape = (50,1), return_sequences=False))
model.add(Dense(1,activation = 'sigmoid'))

model.summary()
```

```python
# model architecture
model2 = Sequential()
model2.add(Embedding(input_dim = 10000, output_dim = 2, input_length = 50))
model2.add(Bidirectional(SimpleRNN(32, input_shape = (50,1), return_sequences=False)))
model2.add(Dense(1,activation = 'sigmoid'))

model2.summary()
```

```python
# model compile
model.compile(metrics = ['acc'], optimizer = 'adam', loss = 'binary_crossentropy')
model2.compile(metrics = ['acc'], optimizer = 'adam', loss = 'binary_crossentropy')
```

```python
# RNN model train
# model Training
model.fit(X_train, y_train, epochs = 5, validation_data = (X_test, y_test))
```

```python
# Bi-RNN model train
# model Training
model2.fit(X_train, y_train, epochs = 5, validation_data = (X_test, y_test))
```

```python
# Prediction
X_test[0][0:50].reshape(1,-1).shape
```

```python
test_data = X_test[0][0:50].reshape(1,-1)
```

```python
# RNN model prediction
if model.predict(test_data) < 0.5:
  print('negative review')
else:
  print('positive review')
```

```python
# bi-RNN model prediction
if model2.predict(test_data) < 0.5:
  print('negative review')
else:
  print('positive review')
```
